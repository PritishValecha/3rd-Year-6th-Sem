```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("covid_vaccine_statewise.csv")
df
```

```python
df.isnull().sum()
```

```python
#dropping sessions and site columns
df.drop("Sessions", axis=1, inplace=True)
df.drop(" Sites ", axis=1, inplace=True)
df
```

```python
count = df['State'].value_counts()['India']
print("Number of occurrences of 'India':", count)
```

```python
#removing rows where state=india
df = df[df['State'] != 'India']
df
```

```python
#selecting necessary columns
x=df.iloc[:,1:5]
x
```

```python
#dropping null values
x.dropna(how="any", inplace=True)
x
```

```python
import matplotlib.pyplot as plt
```

```python
#Bar Plot
vaccination = x.groupby('State')['Total Doses Administered'].sum().reset_index()
vaccination_sort = vaccination.sort_values(by='Total Doses Administered', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(vaccination_sort['State'], vaccination_sort['Total Doses Administered'], color='lightblue')
plt.xlabel('States')
plt.ylabel('Total Doses Administered')
plt.title('Total COVID-19 Vaccination Doses Administered per State')
plt.xticks(rotation=90) 
# plt.tight_layout()
plt.show()
```
